```python
# Importa a biblioteca numpy para operações numéricas, incluindo NaN (Not a Number)
import numpy as np
# Criação de um data frame teste para usar de exemplo na demonstração de preenchimento de nulos.
df_salarios = pd.DataFrame({
   'nome':["Ana", "Bruno", "Carlos", "Daniele", "Val"],
   'salario':[4000, np.nan, 5000, np.nan, 100000]
   })
# Calcula a média salarial e substitui os nulos pela média e arredonda os valores.
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
# Calcula a mediana e substitui os nulos pela mediana.
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
# Exibe o DataFrame de teste com as novas colunas preenchidas
df_salarios
```

```python
# Criação de um data frame teste para usar de exemplo na demonstração de preenchimento de nulos com ffill.
df_temperaturas = pd.DataFrame ({
   'dia':["Segunda", "Terça-Ferira", "Quarta-Feira", "Quinta-Feira","Sexta-Feira"],
   'temperatura':[30, np.nan, np.nan, 28, 27]
    })
# Preenche os valores nulos na coluna 'temperatura' com o último valor válido (forward fill)
df_temperaturas["preenchido_ffill"] = df_temperaturas["temperatura"].ffill()
# Exibe o DataFrame de teste com a nova coluna preenchida
df_temperaturas
```

```python
# Criação de um data frame teste para usar de exemplo na demonstração de preenchimento de nulos com bfill.
df_temperaturas = pd.DataFrame ({
   'dia':["Segunda", "Terça-Ferira", "Quarta-Feira", "Quinta-Feira","Sexta-Feira"],
   'temperatura':[30, np.nan, np.nan, 28, 27]
    })
# Preenche os valores nulos na coluna 'temperatura' com o próximo valor válido (backward fill)
df_temperaturas["preenchido_bfill"] = df_temperaturas["temperatura"].bfill()
# Exibe o DataFrame de teste com a nova coluna preenchida
df_temperaturas
```

```python
# Criação de um data frame teste para usar de exemplo na demonstração de preenchimento de nulos com um valor fixo.
df_cidades = pd.DataFrame({
   'nome':["Ana", "Bruno", "Carlos", "Daniele", "Val"],
   'cidade':["São Paulo", np.nan, "Curitiba", np.nan, "Belém"]
})
# Troca o valor "nulo ou NaN" por "não informado" em cidade_preenchida.
df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("Não Informado")
# Exibe o DataFrame de teste com a nova coluna preenchida
df_cidades
```

```python
from ast import Nonlocal
# Importa a biblioteca pycountry
import pycountry

# Função para converter códigos de país ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar uma nova coluna com código ISO-3 aplicando a função à coluna 'residencia_empregado'
df_limpo['residencia_empregado_iso3'] = df_limpo['residencia_empregado'].apply(iso2_to_iso3)

#Calcular média salarial por país (ISO-3) para o cargo de 'Data Scientist'
# Filtra o DataFrame para incluir apenas 'Data Scientist'
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
# Agrupa por código ISO-3 da residência do empregado e calcula a média do salário em USD
media_ds_pais = df_ds.groupby('residencia_empregado_iso3')['usd'].mean().reset_index()

#Gerar o mapa coroplético
# Cria um mapa coroplético interativo usando Plotly
fig = px.choropleth(media_ds_pais,
                    locations='residencia_empregado_iso3', # Coluna com os códigos de país para mapeamento
                    color='usd', # Coluna usada para colorir o mapa (salário médio)
                    color_continuous_scale='rdylgn', # Escala de cores
                    title='Sálario médio de Cientista de Dados por país', # Título do mapa
                    labels={'usd': 'Salário médio (USD)', 'residencia_empregado_iso3': 'País'}) # Rótulos dos eixos e legenda

# Exibe o mapa
fig.show()
```
